```python
conn = sqlite3.connect('cases.db')
cur = conn.cursor()
```

```python
# read table as pandas dataframe
df = pandas.read_sql_query("SELECT * from CasesVerdict", conn)
```

```python
def get_verdict_head(verdict_text):
    if verdict_text is not None:
        pattern = r'\s'
        head = re.split(pattern, verdict_text)
        return head[0]
    else:
        return None

df['verdict_head'] = df['verdict_text'].progress_apply(get_verdict_head)
```

```python
# check if the verdict_head column is created
def find_max_min_length(df, column_name):
    if column_name not in df.columns:
        return None, None
    
    lengths = df[column_name].astype(str).str.len()
    return lengths.max(), lengths.min()
find_max_min_length(df,'verdict_head')
```

```python
def get_verdict_category(verdict_head):
    if verdict_head is not None:
        if verdict_head[-2:] == '判決':
            return '判決'
        elif verdict_head[-2:] == '裁定':
            return '裁定'
        elif verdict_head[-2:] == '筆錄':
            return '筆錄'
    return None

df['verdict_category'] = df['verdict_head'].progress_apply(get_verdict_category)
```
